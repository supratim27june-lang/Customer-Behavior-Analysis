```python

import pandas as pd
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
df[["Review Rating"]] = imputer.fit_transform(df[["Review Rating"]])
```

```python
df.isnull().sum()
```

```python
df.columns =  df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={"purchase_amount_(usd)": "purchase_amount"})
```

```python
df.columns
```

```python
df.isna().any().any()
```

```python
#creating a column age group
labels = ['Young_adult','Adult','Middle-aged','Senior']
df['age_grp'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df.head(3)
```

```python
is_identical = (df["discount_applied"] == df["promo_code_used"]).all()
print(is_identical)
```

```python
df = df.drop("promo_code_used", axis =1)
```

```python
df.head(3)
```

```python
#creating a mapping of purchases
frequency_mapping = {'Fortnightly': 14, 'Weekly': 7, 'Monthly': 30, 'Quarterly':90, 'Bi-Weekly': 14, 'Annually':365 , 'Every 3 Months': 90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
